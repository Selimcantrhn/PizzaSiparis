```python
import csv
import datetime
# Menu.txt dosyasını oluştur
with open("Menu.txt", "w") as f:
    f.write("* Lütfen Bir Pizza Tabanı Seçiniz:\n1: Klasik\n2: Margarita\n3: TürkPizza\n4: Sade Pizza\n* ve seçeceğiniz sos:\n11: Zeytin\n12: Mantarlar\n13: Keçi Peyniri\n14: Et\n15: Soğan\n16: Mısır\n* Teşekkür ederiz!")



# Pizza üst sınıfı
class Pizza:
    def get_description(self):
        pass

    def get_cost(self):
        pass

# Pizza alt sınıfları
class KlasikPizza(Pizza):
    def __init__(self):
        self.description = "Klasik Pizza"
        self.cost = 15.0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

class MargaritaPizza(Pizza):
    def __init__(self):
        self.description = "Margarita Pizza"
        self.cost = 20.0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

class TurkPizza(Pizza):
    def __init__(self):
        self.description = "Turk Pizza"
        self.cost = 18.0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

class SadePizza(Pizza):
    def __init__(self):
        self.description = "Sade Pizza"
        self.cost = 12.0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

# Decorator üst sınıfı
class Decorator(Pizza):
    def __init__(self, component):
        self.component = component

    def get_description(self):
        pass

    def get_cost(self):
        pass

# Soslar
class Zeytin(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.component = "Zeytin"
        self.cost = 1.5
    def get_description(self):
        return self.component

    def get_cost(self):
        return self.cost

class Mantarlar(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Mantarlar"
        self.cost = 2.0

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.cost

class KeciPeyniri(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Keci Peyniri"
        self.cost = 2.5

    def get_description(self):
        return  self.description

    def get_cost(self):
        return self.cost

class Et(Decorator):
    def __init__(self, component):
        super().__init__(component)
        self.description = "Et"
        self.cost = 3.0

    def  get_description(self):
        return self.description

    def get_cost(self):
        return self.cost
class Sogan(Decorator):
  def init(self, component):
    super().init(component)
    self.description = "Soğan"
    self.cost = 1.0
  def  get_description(self):
        return self.description

  def get_cost(self):
        return self.cost
class Misir(Decorator):
  def init(self, component):
    super().init(component)
    self.description = "Mısır"
    self.cost = 1.0
  def get_description(self):
        return self.description

  def get_cost(self):
        return self.cost
def pChoice():
  global pizza
  pizza_choice = int(input("Lütfen bir pizza tabanı seçiniz: "))
  if pizza_choice == 1:
    pizza = KlasikPizza()
  elif pizza_choice == 2:
    pizza = MargaritaPizza()
  elif pizza_choice == 3:
    pizza = TurkPizza()
  elif pizza_choice == 4:
    pizza = SadePizza()
  else:
    print("Geçersiz seçim!")
  return

def sChoice():
  global decorator  
  sos_choice = input("Hangi sosları eklemek istersiniz? (Örnek: 11, 12): ")
  sos_list = sos_choice.split(",")
  for sos in sos_list:
    if sos == "11":
        decorator = Zeytin(Decorator)
        choosen_sos_cost.append(1.5)
        choosen_sos.append("Zeytin")
    elif sos == "12":
        decorator = Mantarlar(Decorator)
        choosen_sos_cost.append(2.0)
        choosen_sos.append("Mantar")
    elif sos == "13":
        decorator = KeciPeyniri(Decorator)
        choosen_sos_cost.append(2.5)
        choosen_sos.append("Keçi Peyniri")
    elif sos == "14":
        decorator = Et(Decorator)
        choosen_sos_cost.append(3.0)
        choosen_sos.append("Et")
    elif sos == "15":
        decorator = Sogan(Decorator)
        choosen_sos_cost.append(1.0)
        choosen_sos.append("Soğan")
    elif sos == "16":
        decorator = Misir(Decorator)
        choosen_sos_cost.append(1.0)
        choosen_sos.append("Mısır")
    else:
        print("Geçersiz seçim!")
        return
  baska_sos = input("Başka sos eklemek ister misin? (E, H): ")
  if baska_sos == "E":
    sChoice()
  else:
    pass

choosen_sos_cost = []
choosen_sos = []

def allCost():
  print("Toplam tutar: ", pizza.get_cost() + sum(choosen_sos_cost))

def siparis():
# kullanıcı bilgilerini al
    ad = input("Adınız: ")
    tc = input("TC Kimlik Numaranız: ")
    kart_no = input("Kredi Kartı Numaranız: ")
    kart_sifre = input("Kredi Kartı Şifreniz: ")

# sipariş bilgilerini kaydet
    siparis = {
    'Kullanıcı Adı': ad,
    'TC Kimlik Numarası': tc,
    'Kredi Kartı Numarası': kart_no,
    'Sipariş Açıklaması': [pizza.get_description(),choosen_sos],
    'Sipariş Zamanı': datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    'Kredi Kartı Şifresi': kart_sifre
    }
    print(siparis)
    with open('Orders_Database.csv', 'a', newline='') as f:
      writer = csv.DictWriter(f, fieldnames=siparis.keys())
      writer.writerow(siparis)
      choosen_sos_cost.clear()
      choosen_sos.clear()
          
def main():
    # Menu.txt dosyasından menüyü okuyun
  with open('Menu.txt', 'r') as f:
      menu = f.read()

  print(menu)
  pChoice()
  sChoice()
  print("Seçtiğiniz pizza: ", pizza.get_description())
  print(f"Seçtiğiniz soslar: {choosen_sos}")
  allCost()
  siparis()
  
main()
```
